```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

#Seaborn is a framework that simplifies a lot matplotlib functions

# Load the data from the Excel file
df = pd.read_excel("IGN.xlsx")

# Remove any duplicate rows
df.drop_duplicates(inplace=True)

# Remove any games with missing values
df.dropna(inplace=True)

# Convert the "released_date" column to a datetime object
df["released_date"] = pd.to_datetime(df["released_date"], format="%Y-%m-%d")

# Add a "year" column to group the games by year
df["year"] = df["released_date"].dt.year

# Replace any missing values in the "score" column with the mean score
df["score"].fillna(df["score"].mean(), inplace=True)

# Plot a bar chart showing the number of games released per year
plt.figure(figsize=(22, 6))
sns.countplot(x="year", data=df)
plt.xlabel("Year")
plt.ylabel("Number of Games Released")
plt.title("Number of Games Released per Year")
plt.show()

```

2003 was clearly the peak of number of games released

```python
# Add a "year" column to group the games by year
df["year"] = df["released_date"].dt.year

# Calculate the average score for each year
year_avg_score = df.groupby("year")["score"].mean()

# Plot a bar chart showing the average score by year
plt.figure(figsize=(22, 6))
sns.barplot(x=year_avg_score.index, y=year_avg_score.values)
plt.xlabel("Year")
plt.ylabel("Average Score")
plt.title("Average Score by Year")
plt.show()
```

```python
# Add a "year" column to group the games by year
df["year"] = df["released_date"].dt.year

# Group the games by year and find the highest score for each year
year_max_score = df.groupby("year")["score"].max()

# Plot a line chart showing the frequency of the highest score by year
plt.figure(figsize=(8, 6))
sns.lineplot(x=year_max_score.index, y=year_max_score.values)
plt.xlabel("Year")
plt.ylabel("Frequency of Highest Score")
plt.title("Frequency of Highest Score by Year")
plt.show()
```

```python


# Group the games by franchise and find the average score for each franchise
franchise_avg_score = df.groupby("franchises")["score"].mean()



# Filter the franchises by average score greater than or equal to 8 and at least two instances of the franchise
high_score_franchises = franchise_avg_score[(df["franchises"].value_counts() >= 2) & (franchise_avg_score >= 9)].sort_values(ascending=False)

# Get the list of franchises that meet the filtering criteria
high_score_franchise_names = high_score_franchises.index.tolist()

# Filter the original dataframe to only include games from the high score franchises
high_score_games = df[df["franchises"].isin(high_score_franchise_names)]

# Group the games by franchise and count the number of games in each franchise
franchise_counts = high_score_games["franchises"].value_counts()

# Sort the franchises by descending order of average score
sorted_franchises = high_score_franchises.loc[high_score_franchise_names].sort_values(ascending=False)

# Create a line chart showing the average score of each franchise
plt.figure(figsize=(22,6))
for franchise in sorted_franchises.index:
    franchise_games = high_score_games[high_score_games["franchises"] == franchise]
    plt.plot(franchise_games.index, franchise_games["score"], label=franchise)

# Set the x-axis label, y-axis label, and title
plt.xlabel("Game Index")
plt.ylabel("Score")
plt.title("Average Score of High-Scoring Franchises with at least 2 Entries and Average Score > 9 (Descending Order)")

# Show the legend and the plot
plt.legend()
plt.show()
```

```python
# Group the games by franchise and find the average score for each franchise
franchise_avg_score = df.groupby("franchises")["score"].mean()

# Filter the franchises by average score greater than or equal to 8 and at least two instances of the franchise
high_score_franchises = franchise_avg_score[(df["franchises"].value_counts() >= 2) & (franchise_avg_score >= 9)].sort_values(ascending=False)

# Get the list of franchises that meet the filtering criteria
high_score_franchise_names = high_score_franchises.index.tolist()

# Filter the original dataframe to only include games from the high score franchises
high_score_games = df[df["franchises"].isin(high_score_franchise_names)]

# Group the games by franchise and count the number of games in each franchise
franchise_counts = high_score_games["franchises"].value_counts()

# Calculate the total score for each franchise
franchise_total_score = high_score_games.groupby("franchises")["score"].sum()

# Calculate the average score for each franchise
franchise_avg_score = franchise_total_score / franchise_counts

# Sort the franchises by descending order of average score
sorted_franchises = franchise_avg_score.loc[high_score_franchise_names].sort_values(ascending=False)

# Create a bar chart showing the average score of each franchise
plt.figure(figsize=(42,10))
plt.bar(sorted_franchises.index, sorted_franchises.values)

# Set the x-axis label, y-axis label, and title
plt.xlabel("Franchise")
plt.ylabel("Average Score")
plt.title("Average Score of High-Scoring Franchises with at least 2 Entries and Average Score > 9 (Descending Order)")
plt.ylim(8.5, 10)

# Show the plot
plt.show()
```

Bar chart of highest average scoring franchises

```python


# Sort the games by descending order of score and get the top 50
top_games = df.sort_values("score", ascending=False).head(76)

# Create a horizontal bar chart showing the score and release year of each game
plt.figure(figsize=(10,15))
plt.barh(top_games["game"], top_games["score"], color='green')
plt.xlabel("Score")
plt.ylabel("Game")
plt.title("Top 50 Games and Release Year")

# Add the release year as a label on each bar
for i, v in enumerate(top_games["released_date"]):
    plt.text(top_games["score"].iloc[i] + 0.2, i, str(v)[:4], color='black', fontsize=9)

# Set the x-axis limits to start at 9
plt.xlim(9, 10)

# Show the plot
plt.show()
```

Highest average score per game platform

```python
# Calculate the count of games for each platform
platform_counts = df["platform"].value_counts()

# Keep only the platforms with at least 50 games
popular_platforms = platform_counts[platform_counts >= 50].index

# Calculate the mean score for each of these platforms
platform_means = df[df["platform"].isin(popular_platforms)].groupby("platform")["score"].mean()

# Sort the mean scores in descending order
platform_means_sorted = platform_means.sort_values(ascending=False)

# Create a horizontal bar chart of the mean scores
platform_means_sorted.plot(kind="barh")

# Set the title and axis labels
plt.title("Platforms with the Highest Average Game Scores (at least 50 games)")
plt.xlabel("Average Score")
plt.ylabel("Platform")

# Show the chart
plt.show()
```
